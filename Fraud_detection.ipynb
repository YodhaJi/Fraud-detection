```python
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Projects/Fraud detection/Fraud_data.csv")
```

```python
X = data.iloc[:,[1,2,4,5,7,8]]
```

```python
X['type'].value_counts()*100/len(X)
```

```python
y = data.iloc[:,-2].values
```

```python
X.replace("TRANSFER","other",inplace=True)
X.replace("DEBIT","other",inplace=True)
```

```python
X = X.values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [("onehot",OneHotEncoder(),[0])],remainder="passthrough")
X = ct.fit_transform(X)
```

```python
X[0]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [("onehot",OneHotEncoder(),[1,2,8])],remainder="passthrough")
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state =0)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy ='auto',random_state=0)
X_train,y_train = smote.fit_resample(X_train,y_train)
```

```python
X_train[0]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,[4,5,6,7,8]] = sc.fit_transform(X_train[:,[4,5,6,7,8]])
X_test[:,[4,5,6,7,8]] = sc.transform(X_test[:,[4,5,6,7,8]])
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
pred = classifier.predict(X_test)
```

```python
data['isFraud'].value_counts()*100/len(data)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_test,pred))
print(accuracy_score(y_test,pred))
```
